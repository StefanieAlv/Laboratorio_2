GIT es un sistema de control de versiones distribuido de código abierto y gratuito diseñado para manejar todo, desde proyectos pequeños hasta muy grandes, con velocidad y eficiencia. Permite un entorno fluido y colaborativo, es fácil de aprender y ocupa poco espacio con un rendimiento increíblemente rápido. 



Cuenta con ramificaciones locales económicas, convenientes áreas de preparación y múltiples flujos de trabajo. Entre sus múltiples ventajas se encuentran:

- Verificación de autor (por línea de código, programa o actualización)
- Fechas de modificación
- Más fácil de detectar dónde, cómo y quién causó el bloqueo de un programa o línea
- Gestión de código (notificaciones, modificaciones, gestión de versiones, bifurcaciones)
Los sistemas de control de versiones permiten mantener control de archivos y folders, tomar "snapshots" de cada uno de estos. Además mantienen metadata acerca de los cambios realizados a los archivos o folderes, como timestamps de quién realizó cambios y cuándo se realizaron los cambios.
•	Los sistemas de control de versiones son tan útiles que se pueden utilizar básicamente desde tareas muy simples hasta investigaciones completas, ya sea individuales o grupales.

•  Además, permiten regresar a cambios anteriores en los cuales un Unit testing todavía era satisfactorio a través de una búsqueda binaria.

A pesar de que la interfaz de GIT no es tan amistosa y en ocasiones requiere tomar medidas como hacer una copia de seguridad, eliminar el archivo original y trasladar la copia al repositorio, solo se debe comprender la forma en la que se trabajan los datos:

**1. Type blob**
-	El contenido de los bloques de código
-	Array <byte>

**2.	Type tree**
-	Mapping <stray, tree | blob>

**3.	Type commit**
-	Struct E
-	Tiene padres que describe lo que le precedió (uno o múltiples padres)
•	Array <commit>
•	Puede hacer referencia a uno o varios de los padres por los ID
•	Referencia a un árbol (tree)
•	Almacén de direcciones de contenido donde los objetos se direccionan por su hash
-	Contiene metadatos como el autor (string), mensaje (string) y el contenido (snapshot: tree)

**4.	Type object**
-	Blob | tree | commit
-	Content adress store
-	Map <string, object>

**5.	References**
-	Un mapa de un string a otro, está mapeando nombres legibles por humanos.
-	Con references, los nombres legibles por humanos se pueden usar para referirse a snapshots particulares en el historial en lugar de largas cadenas hexadecimales.

NOTA: Función hash: función mágica que toma una gran cantidad de datos y la convierte en una cadena corta
El modelo de git permite también, generar varías líneas de proyecto, en la que es posible la creación de una divergencia entre el estado de un archivo hacia dos ramas distintas en las que se puede trabajar en dos características del mismo archivo de manera congruente. También, estas dos ramas pueden converger en un mismo archivo al finalizar el desarrollo de ambas características.

##**CONCLUSIONES**

1. GIT es una herramienta muy útil para el trabajo colaborativo debido a su facilidad para subir, mantener, ver y editar archivos en línea.
    
2. A pesar de que su interfaz no es la mejor en cuestión de falta de errores, la forma en la que se estructura la información permite un fácil manejo y gestión de los códigos
    
3. GIT es una de las mejores plataformas para el registro de modificaciones, ramificaciones, versiones y actualizaciones de códigos dado que muestra datos sobre el mismo como el autor, fecha de modificación, ID de la persona que hizo la última modificación, etc.
    
4. GIT es excelente para el trabajo en simultáneo, no solo entre desarrolladores distintos, sino también para la creación y modificación simultánea de características para un mismo programa. La creación de ramas y su posterior unificación permite el desarrollo detallado y por separado de soluciones para un mismo programa.
